```python
df=df.drop(columns=["ï»¿"], axis=0)	
```

```python
df.dropna(inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['Total_toxic_replies']= df['Num_toxic_direct_replies']+df['Num_toxic_direct_replies']
df.head()
```

```python
df['Toxic_conversation']= df['Total_toxic_replies'].apply(lambda x :1 if x>0 else 0 )
df.head()
```

```python
# Select features and target
X = df[['Followers', 'Friends', 'Num_tweets', 'Verified', 'Listed_count', 'Location_True',
        'Age', 'Length', 'Num_users', 'Num_author_replies', 'TOXICITY_x', 'Num_author_toxic_replies']]
y = df['Toxic_conversation']
```

```python
print(df.columns.tolist())
```

```python
# If Location is boolean (True/False), convert to int (1/0)
X.loc[:, 'Location_True'] = X['Location_True'].astype(int)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#Applying PCA method 
pca = PCA()
X_pca = pca.fit_transform(X_scaled)


#plot variance 
plt.figure(figsize=(10,6))
plt.plot(np.cumsum(pca.explained_variance_ratio_),marker = 'o')
plt.axhline(y=0.90,color ='y', linestyle= '--')
plt.title("Explianed Variance by PCA Components")
plt.xlabel("Number of componets ")
plt.ylabel("Cummulative Variance")


```

We chose the number of components such that at least 90% of the variance is explained. Based on the plot above, this corresponds to N components (replace N with the appropriate number after running the code).

```python
pca = PCA()
pca.fit(X_scaled)
explained_var = pca.explained_variance_ratio_
cumulative_var = np.cumsum(explained_var)

# Display components 1 to 10
print("Component | Explained Variance | Cumulative Variance")
print("---------------------------------------------------")
for i in range(10):
    print(f"{i+1:9} | {explained_var[i]:18.4f} | {cumulative_var[i]:19.4f}")
```

```python
pca_final = PCA(n_components=9)
X_pca_final = pca_final.fit_transform(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca_viz[:, 0], X_pca_viz[:, 1], c=y, cmap='coolwarm', alpha=0.6)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA Visualization (2D)')
plt.show()
```

```python
# Split dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_pca_final, y, test_size=0.2, random_state=42)

# Train logistic regression using 5-fold cross-validation
log_reg = LogisticRegression(max_iter=1000)
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

f1_scores = cross_val_score(log_reg, X_train, y_train, scoring='f1', cv=cv)
accuracy_scores = cross_val_score(log_reg, X_train, y_train, scoring='accuracy', cv=cv)

print("Average F1 Score (5-fold CV):", f1_scores.mean())
print("Average Accuracy (5-fold CV):", accuracy_scores.mean())
```

```python
# Fit final model and evaluate on test set
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)

print("Test F1 Score:", f1_score(y_test, y_pred))
print("Test Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Elbow method for choosing k
inertia = []
K_range = range(1, 11)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(K_range, inertia, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid()
plt.show()
```

```python
# Choose k based on elbow method, e.g., k = 4
k_opt = 4
kmeans_final = KMeans(n_clusters=k_opt, random_state=42)
df['Cluster'] = kmeans_final.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(X_pca_viz[:, 0], X_pca_viz[:, 1], c=df['Cluster'], cmap='Set1', alpha=0.6)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title(f'K-Means Clustering with k={k_opt}')
plt.colorbar(label='Cluster')
plt.show()
```
